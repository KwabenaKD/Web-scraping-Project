```python
import time 
import datetime
import pandas as pd 

ticker = 'TSLA'
period1 = int(time.mktime(datetime.datetime(2020,1,1,12,15).timetuple()))
period2 = int(time.mktime(datetime.datetime(2023,1,8,12,15).timetuple()))
interval = '1wk'

query_string = f'https://query1.finance.yahoo.com/v7/finance/download/{ticker}?period1={period1}&period2={period2}&interval={interval}&events=history&includeAdjustedClose=true'

```

```python
df = pd.read_csv(query_string)
```
